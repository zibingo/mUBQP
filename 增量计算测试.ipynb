```python
# -*- encoding: utf-8 -*-
'''
@Author                :   zibin
@Last Modified time    :   2022/05/10 19:46:17
@Contact               :   zibingo@qq.com
'''
import numpy as np
import sys
q1 = [
    [1,2,3],
    [2,1,2],
    [3,1,2]]
q2 = [
    [2,1,3],
    [3,1,1],
    [1,2,3]
]

def Matrix_X(X, q1, q2):
    sum1, sum2 = 0, 0
    n = len(X)  # n*n的矩阵
    for i in range(n):
        for j in range(n):
            sum1 += q1[i][j] * X[i] * X[j]
            sum2 += q2[i][j] * X[i] * X[j]
    return sum1, sum2

def Formula(FlipIndex,X,q):
    sum = 0
    n = len(X)
    for j in range(n):
        if j != FlipIndex:
            sum += 2 * q[FlipIndex][j]* X[j]
    return (1 - 2 * X[FlipIndex]) * (sum  + q[FlipIndex][FlipIndex])

i = 1
X = [0,0,1]
f1For,f2For = Matrix_X(X,q1,q2)
f1For += Formula(i,X,q1)
print("公式翻转后的f1:",f1For)

def InitDeltaC(X,q1,q2):
    DeltaC1,DeltaC2 = [],[]
    n = len(X)
    for j in range(n):
        DeltaC1.append(Formula(j,X,q1))
    for j in range(n):
        DeltaC2.append(Formula(j,X,q2))
    return DeltaC1,DeltaC2

def UpdateDeltaC(X,j,q1,q2,DeltaC1,DeltaC2):
    n = len(X)
    for i in range(n):
        if i == j:
            DeltaC1[i] = - DeltaC1[i]
        else:
            DeltaC1[i] -= (q1[i][j] + q1[j][i]) * (1 - 2 * X[i]) * (1 - 2 * X[j])

    for i in range(n):
        if i == j:
            DeltaC2[i] = - DeltaC2[i]
        else:
            DeltaC2[i] -= (q2[i][j] + q2[j][i]) * (1 - 2 * X[i]) * (1 - 2 * X[j])

DeltaC1,DeltaC2 = InitDeltaC(X,q1,q2)
f1C,f2C = Matrix_X(X,q1,q2)
f1C += DeltaC1[i]
print("记录C翻转后的f1:",f1C)
X[i] = 1 - X[i]
UpdateDeltaC(X,i,q1,q2,DeltaC1,DeltaC2)


#第二次
r = 2
f1For,f2For = Matrix_X(X,q1,q2)
f1For += Formula(r,X,q1)
print("公式翻转后的f1:",f1For)
f1C += DeltaC1[r]
print("记录C翻转后的f1:",f1C)
X[r] = 1 - X[r]
UpdateDeltaC(X,r,q1,q2,DeltaC1,DeltaC2)


```
